```python
#  Исходя из графиков распределения и таблицы статических показателей,видим, что в некоторых
# переменных (R&D Spend и Marketing Spend) присутствуют выбросы
# "Загоним" все значения в три сигма
# Function Outliers
def outliers(df):
    num_var = list(df._get_numeric_data().columns)
    for RD_spend in num_var:
        df[RD_spend] = df[RD_spend].apply(lambda y: df[RD_spend].mean()-3*df[RD_spend].std() 
                            if y < df[RD_spend].mean()-3*df[RD_spend].std() else y)
        df[RD_spend] = df[RD_spend].apply(lambda y: df[RD_spend].mean()+3*df[RD_spend].std() 
                            if y > df[RD_spend].mean()+3*df[RD_spend].std() else y)
    for Marketing_Spend in num_var:
        df[Marketing_Spend] = df[Marketing_Spend].apply(lambda y: df[Marketing_Spend].mean()-3*df[Marketing_Spend].std() 
                            if y < df[Marketing_Spend].mean()-3*df[Marketing_Spend].std() else y)
        df[Marketing_Spend] = df[Marketing_Spend].apply(lambda y: df[Marketing_Spend].mean()+3*df[Marketing_Spend].std() 
                            if y > df[Marketing_Spend].mean()+3*df[Marketing_Spend].std() else y) 
            
    return(df)
```

```python
# Outliers
df = outliers(df)
df.describe()
```
